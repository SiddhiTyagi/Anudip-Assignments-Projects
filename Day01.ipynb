# A company wants to identify whether an employee is eligible for a bonus based on attendance percentage and performance rating

```python
attendance_percentage = float(input("Enter employee's attendance percentage: "))
performance_rating = float(input("Enter employee's performance rating: "))
```

```python
if attendance_percentage >= 75 and performance_rating >= 4:
  print("Employee is eligible for bonus.")
else:
  print("Employee is not eligible for bonus.")
```
