# A company wants to identify whether an employee is eligible for a bonus based on attendance percentage and performance rating

```python
employee_name = input("Enter Employee's Name: ").upper()

attendance_percentage = float(input("Enter "+employee_name+"'s Attendance Percentage: "))
performance_rating = float(input("Enter "+employee_name+"'s Performace Rating out of 5: "))

if attendance_percentage >= 80 and performance_rating >= 4:
    print(employee_name,"is eligible for bonus!")
else:
    print(employee_name,"is NOT eligible for bonus!")
```

# Create a program to check whether a given username and password match predefined credentials

# A traffic monitoring system needs to determine whether a vehicle exceeded the speed limit

```python
speed = int(input("Enter the speed of vehicle (km/h): "))
speed_threshold = 40
difference = speed - speed_threshold

if speed > speed_threshold:
  print(f"Speed limit exceeded by {difference} km/h")
else:
  print("Speed limit NOT exceeded.")
```
