# LEVEL 1: EASY

```python
"""Write a program to check whether a number is positive, negative, or zero."""

try:
  num=float(input("Enter number: ")) # what if this input statement is outside(above) try block?
  if num>0:
    print(f"{num:.2f} is a positive number.")
  elif num<0:
    print(f"{num:.2f} is a negative number.")
  else:
    print("0 is equal to zero.") # I WANT 0 AS 0, NOT AS 0.0

except ValueError:
  print("INVALID INPUT")
```

```python
"""Ask the user to enter 5 numbers and print:
Sum
Average"""

number_of_nums=5
num=[0]*number_of_nums
total=0

for i in range(number_of_nums):
  num[i]=float(input(f"Enter number {i+1}: "))
  total=total+num[i]

average=total/number_of_nums

# print using only 1 print statement
print(f"Sum: {total:.2f}\nAverage: {average:.2f}")
```

```python
"""--------------------Print all numbers between 1 and 100 that are divisible by both 3 and 5."""

print("All numbers between 1 and 100 that are divisible by both 3 and 5 are ",end="")

divisible_num=[]

for i in range(1,101):
  if i%3==0 and i%5==0:
    divisible_num.append(i)

for i in range(len(divisible_num)):
  if i==len(divisible_num)-1:
    print(divisible_num[i])
  else:
    print(divisible_num[i],end=", ")
```

```python
"""Ask the user to enter a string and print:
Length of the string
String in uppercase
String in lowercase"""

string=input("Enter string: ")
print(f"String: {string}")
print("Length of string:",len(string))
print(f"{string} in uppercase:",string.upper())
print(f"{string} in lowercase:",string.lower())
```

```python
"""Check whether a string is empty.
Example:
Input: ""

Output: String is empty."""

string=input("Enter string: ")

if len(string)==0:
  print("String is empty.")
else:
  print("String is NOT empty.")
```

```python
"""Count the total number of characters in a string without using len()."""

string=input("Enter string: ")
count=0

for char in string:
  count+=1

print(f"Total number of characters in {string} is",count)
```

```python
"""Count the number of vowels"""

string=input("Enter string: ")
count=0

for char in string:
  if char.lower() in "aeiou":
    count+=1

print(f"Number of Vowels: {count}")
```

```python
"""Count the number of consonants."""

string=input("Enter string: ")
consonant_count=0

for char in string:
  if char.lower() not in "aeiou":
    consonant_count+=1

print(f"Number of consonants: {consonant_count}")
```

```python
 """Count:
 Uppercase letters
 Lowercase letters
 Digits
 Spaces
 Example: Hello 123
 Output:
 String: Hello 123
 Uppercase letters: 1
 Lowercase letters: 4
 Digits: 3
 Spaces: 1
 """

 string = input("Enter string: ")
 uppercase_count=0
 lowercase_count=0
 digit_count=0
 space_count=0

 for char in string:
  if char.isupper():
    uppercase_count+=1
  elif char.islower():
    lowercase_count+=1
  elif char.isdigit():
    digit_count+=1
  elif char.isspace():
    space_count+=1

 print(f"String: {string}")
 print(f"Uppercase letters: {uppercase_count}")
 print(f"Lowercase letters: {lowercase_count}")
 print(f"Digits: {digit_count}")
 print(f"Spaces: {space_count}")
```

```python
"""Print the first and last character of a string.
Example: Python

First: P
Last: n
"""

string=input("Enter string: ")
string_length=len(string)

for char in range(string_length):
  if char==0:
    print(f"First character: {string[char]}")
  if char==string_length-1:
    print(f"Last character: {string[char]}")
```

```python
"""----------------------------Check whether a particular character exists in a string.
Example:
String: Python

Character: h

Output: Character Found
"""

string=input("Enter string: ")
character=input("Enter character to be found: ")
frequency=0

for char in string:
  if char==character:
    frequency+=1
    if frequency==1:
      print(f"{frequency} Character Found")
    elif frequency>=2:
      print(f"{frequency} Characters Found")
  elif frequency==0:
    print(f"No {character} Character Found in {string}")
```

```python
"""1. Count only alphabets  2. Count only digits
Input: AB2026siddhi @#
Output: Alphabets: 8
Digits: 4
"""

string=input("Enter string: ")
count_alpha=0
count_digit=0

for char in string:
  if char.isalpha():
    count_alpha+=1
  elif char.isdigit():
    count_digit+=1

print(f"Alphabets: {count_alpha}")
print(f"Digits: {count_digit}")
```
