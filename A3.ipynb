# A school wants to calculate grades for students based on marks obtained in five subjects.

```python
student_name = input("Enter the name of Student: ").strip().upper()
english_mark = float(input("Enter the marks of English: "))
maths_mark = float(input("Enter the marks of Maths: "))
physics_mark = float(input("Enter the marks of Physics: "))
biology_mark = float(input("Enter the marks of Biology: "))
chemistry_mark = float(input("Enter the marks of Chemistry: "))

average = (english_mark + maths_mark + physics_mark + biology_mark + chemistry_mark)/5
print(f"Student's Name: {student_name}")
print(f"Average Marks: {average:.2f}")

if average >= 95:
  print("GRADE: A+")
elif average >= 90:
  print("GRADE: A")
elif average >= 85:
  print("GRADE: B+")
elif average >= 80:
  print("GRADE: B")
elif average >= 75:
  print("GRADE: C+")
elif average >= 70:
  print("GRADE: C")
elif average >= 65:
  print("GRADE: D+")
elif average >= 60:
  print("GRADE: D")
else:
  print("GRADE: FAIL")
```

# Create a login attempt system that allows the user a maximum of 3 attempts using a while loop.

```python
correct_username="greekgod1"
correct_password="Stephen@19Dimoldenberg"

attempts=0
max_attempts=3

while attempts<max_attempts:
  username=input("Enter username: ").strip().lower()
  """what about the egde case where the user enters username="greek god ", where there are two space, including a trailing space?
  Do we replace the leading or trailing space to an underscore?
  THE REAL STANDARD: Companies treat leading and trailing spaces as mistakes"""
  password=input("Enter password: ")
  print()

  username=username.replace(" ","_") # optional

  if correct_username==username and correct_password==password: # use tuple way, if (correct_username,correct_password)==(username,password):
    print("LOGGED IN SUCCESSFULLY!")
    break

  else:
    attempts+=1
    attempts_left=max_attempts-attempts
    print("Invalid Username or Password! Attempts left: ",attempts_left)
    print()

if attempts==max_attempts:
  print("MAXIMUM ATTEMPTS ACHIEVED!")
```

# A food delivery app wants to display menu items repeatedly until the user selects Exit.
