# A bank wants to calculate simple interest for customers based on principle, rate and time, entered by the user.

```python
principle = float(input("Enter Principle: "))
rate = float(input("Enter Rate: "))
time = float(input("Enter Time: "))

simple_interest = (principle*rate*time)/100

print(simple_interest)
```

# A movie theatre wants to assign ticket pricing based on age categories such as child, adult and senior citizen.

```python
customer_name = input("Enter Customer's Name: ").upper()
customer_age = int(input(f"Enter {customer_name}'s Age (in years): "))

if customer_age < 18:
  print("Ticket Price is ₹70")
elif customer_age <= 75:
  print("Ticket Price is ₹120")
else:
  print("Ticket Price is ₹150")
```

# Write a program to find the largest among three entered sales figures.

```python
fig1 = float(input("Enter the first Sales Figure (in ₹): "))
fig2 = float(input("Enter the second Sales Figure (in ₹): "))
fig3 = float(input("Enter the third Sales Figure (in ₹): "))

if fig1 > fig2 and fig1 > fig3:
  print(f"₹{fig1} is the largest Sales Figure amongst the three!")
elif fig2 > fig3:
  print(f"₹{fig2} is the largest Sales Figure amongst the three!")
else:
  print(f"₹{fig3} is the largest Sales Figure amongst the three!")
```
